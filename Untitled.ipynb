## looking at diet
Looking at the diet feature alone, there seems to be a pretty consistent trend.

```python
df["diet"].plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel("year", fontsize=20);
```

At the beginning of each year (January), there's a huge uptick in *diet*'s search popularity. This could be directly correlated to New Year's resolutions. Similarly, in the last quarter of each year, there is a significant decrease in popularity up until the new year. Although it may not be relative insight for all countries, in the US this could be driven by the fact that October, November, and December are holiday months, centered around food-centric holidays including, but not limited to, Halloween, Thanksgiving, and Christmas.

### identifying trends
Clearly there seems to be a trend here. We can use the *rolling average* method to smooth out the sharpness of the above curve. This method, to get the value for event at time t, takes the average of points on the following intervals [t-i, t) and (t, t+i], for any real number i. This interval is called a window, and i is the window size.

Below we'll use Pandas's built in functionality with a window size of 12 (for a 12 month or 1 year period leftward/rightward).

```python
diet = df[["diet"]]
diet.head()
```

```python
diet_window_size0 = 12
```

```python
#diet.rolling(12).mean().head() shows NaN because the initial dataframe
#has no leftward neighbors with respect to time
diet.rolling(diet_window_size0).mean()[12:17]
```

```python
diet.rolling(diet_window_size0).mean().plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel("year", fontsize=20);
```

Based on the above visual, it seems `window = 12` is a poor parameter value choice (seems to lose scope of the monthly trends). However, the tutorial uses it for now so we'll stick with it.

### gym
Let's look at the data trends using the *rolling average* method for gym as well.

```python
gym = df["gym"]
gym.plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel("year", fontsize=20)
```

```python
gym.rolling(12).mean().plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel("year", fontsize=20)
```

### thoughts on rolling average
Okay. It seems as though the *rolling average* method is used to remove noise up to the point to see the general trend of the data as opposed to seeing it at a clearer level for the micro picture. This make sense. Gym has become mroe trending in 2018 in comparison to 2005.

## comparing diet and gym trends

```python
df_rm = pd.concat([diet.rolling(diet_window_size0).mean(), gym.rolling(diet_window_size0).mean()], axis=1)
df_rm.plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel("year", fontsize=20)
```

## seasonal patterns
In order to observe the smaller, seasonal trends, it would help to remove the larger overarching trends from the data. One method of doing this is called first-order differencing, which is done on a per-point basis. First-order differencing examines the difference between successive data points.

### first-order differencing plot

```python
diet.diff(periods=1).plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel("year", fontsize=20);
```

We've now really diminished the affect of the trend on the data, and are able to see the pronounced peaks at the beginning of each year more clearly.
