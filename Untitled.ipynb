## imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

## data
Below we'll grab the data. It should be located in the same folder as this notebook.

```python
#the first row is some text from the google analytics download
df = pd.read_csv("multiTimeline.csv", skiprows=2)
df.head(10)
```

```python
df.info()
```

### renaming columns
This is more of just a nicety, but removing the whitespace generally makes things easier to work with.

```python
df.columns.tolist()
```

```python
df.columns = list(map(lambda e: e.split(":")[0].lower(), df.columns.tolist()))
df.columns
```

```python
df.head(10)
```

### converting the `month` column
The month column, as seen in the `.info()` call, is of data type `object`. We'll convert this column specifically to a datetime type.

```python
df.columns
```

```python
#note: make sure to only run this cell once...
#running a second time will raise an error
df["month"] = pd.to_datetime(arg=df["month"])
df.set_index("month", inplace=True)
df.head(10)
```
